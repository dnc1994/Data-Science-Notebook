We first need to import some of the most common tools for analysis, pandas and matplotlib. One thing to note is that setting matplotlib to provide the plot output to the notebook is handled automatically, so you don't have to use `%matplotlib inline`.

Good practice is to rename module imports to a short name, vice importing everything with `*`.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Pandas DataFrame

A dataframe is a table-like structure with labeled arrays. This is a similar concept to a dict, where the *keys* are the labels, and the *values* are the arrays. Creating a dict in this way is an easy method to quickly create a dataframe.

We create x values, y values that square each value in the x values, then create the dataframe.

```python
xs = range(1,50)
ys = [x**2 for x in xs]
df = pd.DataFrame({'x': xs, 'y': ys})
```

An alternative way would be to construct it as follows:

```python
df = pd.DataFrame()
df # empty
```

```python
# note: cannot use dot notation if the column isn't already in the dataframe
df['x'] = range(1,50)
df['y'] = df['x']**2
```

Below is an example of how you can peek into the dataframe contents and get a nice table view of the structure.

```python
df.head()
```

### Plotting

The dataframe provide dot (.) access into the arrays that it contains. With matplotlib, which we renamed to `plt`, we can quickly plot y versus x.

```python
p = plt.plot(df.x, df.y)
```

### Filtering

The benefit of the dataframe is the ease in logical filtering of the rows by some conditional expression.

```python
# select only a portion of our data frame where x < 20 or x > 30
df_filt = df[(df['x'] < 20) | (df['x'] > 30)]

# use filtered dataframe for plotting
plt.plot(df_filt.x, df_filt.y, 'bo')
```
